Applying Linear Regression in datasets. Data set from https://archive.ics.uci.edu/ml/machine-learning-databases/housing/

# 0. Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 1. Read the the dataset

```python
DATA_CSV_FILE = pd.read_csv('Salary.csv')
```

```python
DATA_CSV_FILE.isnull().sum()
```

# 2. Exploratory Data Analysis

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.scatterplot(data=DATA_CSV_FILE, x="Salary", y="YearsExperience")
sns.lineplot(data=DATA_CSV_FILE, x="Salary", y="YearsExperience")
plt.show()
```

# 3. Preparing Data for training and testing

```python
X = pd.DataFrame(np.c_[DATA_CSV_FILE['YearsExperience']], columns=['YearsExperience'])
Y = DATA_CSV_FILE['Salary']

from sklearn.model_selection import train_test_split
```

# 4. Training and Testing the model

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

predict_salary = lin_model.predict([[12],[10],[17.8],[20],[3], [0.1]])
print(predict_salary)
print(lin_model.score(X_test, y_test))
```
